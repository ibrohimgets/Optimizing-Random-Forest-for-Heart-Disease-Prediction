## Option 2 fill the missing datas with SKlearn

```python
missing_car = pd.read_csv("data/car-sales-extended-missing-data.csv")
missing_car.head()
```

```python
missing_car.isna().sum()
```

```python
missing_car.dropna(subset = ["Price"], inplace=True)
missing_car.isna().sum()
```

```python
x = missing_car.drop("Price", axis = 1)
y = missing_car["Price"]
```

# Fill the missing data with Sklearn

```python
from sklearn.impute import SimpleImputer 
from sklearn.compose import ColumnTransformer 

cat_imputer = SimpleImputer(strategy = "constant", fill_value = "missing")
door_imputer = SimpleImputer(strategy = "constant", fill_value = 4)
number_imputer = SimpleImputer(strategy = "mean")

#Define the columns 
cat_features = ["Make", "Colour"]
door_features = ["Doors"]
number_features = ["Odometer (KM)"]

#Create an imputer 
imputer = ColumnTransformer([
    ("cat_imputer", cat_imputer, cat_features),
    ("door_imputer", door_imputer, door_features),
    ("number_imputer", number_imputer, number_features),
])
# Transform the data 
filled_x = imputer.fit_transform(x)
filled_x
```

```python
car_sales_filled = pd.DataFrame(filled_x,
                               columns=["Make", "Colour", "Doors", "Odometer (KM)"])
car_sales_filled
```

```python
car_sales_filled.isna().sum()
```

# Let's now convert the filled data into the number again!

```python
columns =  ["Make", "Colour", "Doors"]
one_hot = OneHotEncoder()
convert = ColumnTransformer([("new_column", 
                                 one_hot,
                                 columns)], 
                                remainder="passthrough")
new_x = convert.fit_transform(car_sales_filled)
new_x
```

# We got all the data in numbers now! Let's fit and test the model!

```python
np.random.seed(42)

x_train, x_test, y_train, y_test = train_test_split(new_x, 
                                                    y, 
                                                    test_size=0.2, 
                                                    )
model = RandomForestRegressor(n_estimators=100)
model.fit(x_train, y_train)
model.score(x_test, y_test)
```
